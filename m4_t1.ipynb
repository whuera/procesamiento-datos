```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python

Ventas_2021 = pd.DataFrame({"Clientes": ("Juan", "Camila", "Daniel", "Luis", "Juan", "Ana", "Camila", "Ana","Luis", "Juan", "Ana", "Juan"),
                            "Ventas":(100.0,200.0,110.00,90.0,150.0,80.0,70.0,100.0,100.0,110.0,90.0,100.00)})
Ventas_2021
```

```python
eje_x = ["Ana", "Camila", "Daniel", "Juan", "Luis"]
eje_y = 270.0, 270.0, 110.0, 460.0, 190.0
plt.bar(eje_x, eje_y, color = "lightblue", label = "Ventas")
plt.xlabel("Clientes")
plt.ylabel("Ventas 2021")
plt.title("Reporte de Ventas")
plt.show()
```

```python
Ventas_2021.describe()
```

```python
media_ventas = Ventas_2021["Ventas"].mean()
min_ventas = Ventas_2021["Ventas"].min()
max_ventas = Ventas_2021["Ventas"].max()
std_ventas = Ventas_2021["Ventas"].std()
print("Media de ventas:", media_ventas)
print("Mínimo de ventas:", min_ventas)
print("Máximo de ventas:", max_ventas)
print("Desviación estándar de ventas:", std_ventas)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generar el diagrama de caja y bigote (boxplot)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.boxplot(x="Clientes", y="Ventas", data=Ventas_2021)
plt.xlabel("Clientes")
plt.ylabel("Ventas")
plt.title("Diagrama de Caja y Bigote de Ventas 2021")
plt.show()
```
