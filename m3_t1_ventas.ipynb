```python
ventas_clientes_2021 = pd.Series([100.0,200.0,110.00,90.0,150.0,80.0,70.0,100.0,100.0,110.0,90.0,110.00])
ventas_clientes_2021
Clientes_2021 = pd.Series(["Juan","Camila","Daniel","Luis","Juan","Ana","Camila","Ana","Luis","Juan","Ana","Juan"])
Clientes_2021
df1 = pd.DataFrame(ventas_clientes_2021, columns=["Ventas"])
df2 = pd.DataFrame(Clientes_2021, columns=["Clientes"])
df3 = pd.concat([df2,df1], axis = 1)
df3
df3.sort_values(by=['Clientes'], ascending = True)
```

```python
# Agrupar por Producto y Vendedor y sumar las ventas
ventas_por_grupo = df3.groupby(['Clientes','Ventas'])['Ventas'].sum()
ventas_por_grupo

```

**Ordenar y Agrupar Datos**

```python
Clientes_2021 = pd.Series(["JUan","camila","DanIel","Luis","Juan","Ana","Camila","Ana","LuiS","Juan","Ana","Juan"])
Clientes_2021 = Clientes_2021.str.capitalize()
df1 = pd.DataFrame(ventas_clientes_2021, columns=["Ventas"])
df2 = pd.DataFrame(Clientes_2021, columns=["Clientes"])
df3 = pd.concat([df2,df1], axis = 1)
df3
df3.sort_values(by=['Clientes'], ascending = True)
ventas_por_grupo1 = df3.groupby(['Clientes','Ventas'], as_index=False).agg({'Ventas': 'sum'})
ventas_por_grupo1
ventas_por_grupo1.groupby(['Clientes','Ventas']).agg(total=('Ventas', min))
ventas_por_grupo1
```

**Ordenar y procesar datos**

```python
compute_ventas = ventas_por_grupo1.pivot_table("Ventas", index = "Clientes", aggfunc = {"Ventas": ["count", sum, "mean"]})
df_compute = pd.DataFrame(compute_ventas)
df_compute
```

**Obtener la diferencia de ventas**

```python
valor_max = df_compute.max()
valor_max
valor_min = df_compute.min()
valor_min
val_dif = valor_max - valor_min
val_dif
```
